```r
# Load necessary libraries
# Load necessary libraries
library(data.table)
library(tidyverse)

# Function to perform QC on SNPs
qc_snps <- function(snp_locations, geno_mat) {
  # Convert geno_mat to data.table if it's not already
  setDT(geno_mat)
  
  # Remove SNPs with NA values
  geno_mat <- geno_mat[complete.cases(geno_mat)]
  
  # Function to filter SNPs with at least 3 individuals per genotypic category
  filter_snps <- function(geno_mat) {
    # Initialize a vector to store valid SNPs
    valid_snps <- vector("logical", nrow(geno_mat))
    
    # Iterate over each row (SNP) in geno_mat
    for (i in seq_len(nrow(geno_mat))) {
      snp_row <- geno_mat[i, -1, with = FALSE]  # Exclude the 'snp' column
      counts <- table(unlist(snp_row))
      
      # Check if there are at least 3 individuals per genotypic category (0, 1, 2)
      if (all(counts[c("0", "1", "2")] >= 3, na.rm = TRUE)) {
        valid_snps[i] <- TRUE
      } else {
        valid_snps[i] <- FALSE
      }
    }
    
    # Return the SNPs that are valid
    return(geno_mat[valid_snps, snp])
  }
  
  # Filter SNPs
  valid_snps <- filter_snps(geno_mat)
  geno_mat_filtered <- geno_mat[snp %in% valid_snps]
  
  # Merge with snp_locations to keep only valid SNPs
  snp_locations_filtered <- snp_locations[annot %in% valid_snps]
  
  # Summary
  summary <- list(
    total_snps = nrow(snp_locations),
    valid_snps = length(valid_snps),
    removed_snps = nrow(snp_locations) - length(valid_snps)
  )
  
  return(list(
    geno_mat_filtered = geno_mat_filtered,
    snp_locations_filtered = snp_locations_filtered,
    summary = summary
  ))
}

# Example usage
# Assuming snp_locations and geno_mat are already loaded data frames
result <- qc_snps(snp_locations, geno_mat)

# Access the filtered data frames and summary
geno_mat_filtered <- result$geno_mat_filtered
snp_locations_filtered <- result$snp_locations_filtered
summary <- result$summary

# Print summary
print(summary)
```
